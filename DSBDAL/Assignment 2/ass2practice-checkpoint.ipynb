```python
import numpy as np

# Convert 'math score' column to numeric
df['math score'] = pd.to_numeric(df['math score'], errors='coerce')

# Calculate quartiles and IQR
Q1 = np.percentile(df['math score'].dropna(), 25)
Q3 = np.percentile(df['math score'].dropna(), 75 )
IQR = Q3 - Q1
print(IQR)
print(Q1)
print(Q3)

# Calculate upper and lower bounds for outlier removal
upper = Q3 + 1.5 * IQR
lower = Q1 - 1.5 * IQR
print(upper)
print(lower)

# Filter DataFrame to remove outliers
df = df[(df['math score'] >= lower) & (df['math score'] <= upper)]
```

```python
sns.boxplot(x='math score',data=df)
```
